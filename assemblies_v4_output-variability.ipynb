```python
def calc_error(x, y):
    x_err = []
    for i in range(exp.net.p.trials):
        # Estimate
        model = sm.OLS(y, x[i].T)
        params = model.fit().params

        # Predict
        ye = np.dot(x[i].T, params)
        err = np.sqrt(np.sum(np.square(y-ye)))
        x_err.append(err)
    return x_err
```

```python
plt.figure(figsize=(10,3))
plt.plot(calc_error(a_x, a_y), label='A')
plt.plot(calc_error(b_x, b_y), label='B')
plt.xlim(0,39)
plt.xlabel('trials')
plt.ylabel('RMSE')
plt.legend()
pl = plt.show()
```
